# Projeto de Visualização da Informação

## Introdução

Neste projeto foi utilizado um dataset real contendo o histórico da cotação do dólar em relação ao real (USD/BRL).

O objetivo do trabalho foi criar diferentes tipos de visualizações para analisar o comportamento dos dados ao longo do tempo.

As bibliotecas utilizadas foram:
- pandas
- matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("USD_BRL_hist.csv")

df["Data"] = pd.to_datetime(df["Data"], format="%d.%m.%Y")

df = df.sort_values("Data")

df["Mes"] = df["Data"].dt.to_period("M").astype(str)

media_mensal = df.groupby("Mes")["USD_BRL"].mean().reset_index()

df["Indice"] = range(len(df))
```
